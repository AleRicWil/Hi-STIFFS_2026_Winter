```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib tk
from sympy.physics.vector.printing import vlatex
from IPython.display import Math, display

sp.init_session()
sp.init_printing()

def dotprint1(expr):
    display(Math(vlatex(expr)))

def dotprint(expr, lineName=0):
    if lineName == 0:
        if isinstance(expr, tuple): # If expr is a tuple, wrap each element in square brackets
            expr = [f'[{vlatex(e)}]' for e in expr]
        else: # If expr is a single element, no square brackets
            expr = [vlatex(expr)]
        display(Math(r'\;'.join(expr)))
    else:
        # Replace spaces in lineName with `\;` for LaTeX formatting and place it at the beginning
        formatted_lineName = r"\;".join(lineName.split())
        if isinstance(expr, tuple): # If expr is a tuple, wrap each element in square brackets
            expr = [f'[{vlatex(e)}]' for e in expr]
        else: # If expr is a single element, no square brackets
            expr = [vlatex(expr)]
        # Only add formatted_lineName at the start, not before each element of expr
        display(Math(f'{formatted_lineName}:\; ' + r',\;\;'.join(expr)))
```
